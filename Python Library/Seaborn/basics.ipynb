```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
# basic plottinf with seaborn
df = sns.load_dataset('tips')
print(df)
```

```python
# create scatter plo
# t
sns.scatterplot(x='total_bill', y ='tip', data=df)
plt.title('scatter plot of total bill vs tip')
```

```python
plt.scatter(df['total_bill'], df['tip'])
```

```python
# line plot
sns.lineplot(x ='size',y='total_bill', data=df)
plt.title('line plot of size vs total bill')
plt.show()
```

```python
# categorical plot
# bar plot
sns.barplot(x='day', y='total_bill', data=df)
```

```python
# box plot
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
# violin plot
sns.violinplot(x='day', y='total_bill', data=df) 
```

```python
# histogram 
sns.histplot(df['total_bill'],bins = 10,kde=True)
```

```python
# kde plot
sns.kdeplot(df['total_bill'],fill=True)
```

```python
# pairplot
sns.pairplot(df)
```

```python
sns.pairplot(df)
```

```python
corr = df[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True)
```

```python
sns.heatmap(corr, annot=True,cmap='coolwarm')
```
