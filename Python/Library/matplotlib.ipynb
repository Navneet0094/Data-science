```python
import matplotlib.pyplot as plt
```

```python
 x = [1, 2, 3, 4, 5]
 y = [1, 4, 9, 16, 25]
 plt.plot(x,y)
 plt.xlabel('x axis label')
 plt.ylabel('y axis label')
 plt.title('Simple plot')
 plt.show()
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=4,markersize=12)
plt.grid(True)
```
