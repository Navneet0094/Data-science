#### Stemming

```python
''' 
Classification problem 
Comments of product is a positive review or negative review
Reviews -> eaten , eating ,eat or [going , gone ,goes]
'''
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
## Porter Stemmer
from nltk.stem import PorterStemmer
stemming = PorterStemmer() 
```

```python
for i in words:
    print(i," ---> ",stemming.stem(i))
```

```python
word = ['congratulations', 'sitting']
stems = [print(w," --> ",stemming.stem(w)) for w in word]
# no proper meaning 
```

```python
# RegexpStemmer
from nltk.stem import RegexpStemmer
```

```python
rstemming = RegexpStemmer('ing$|s$|e$|able$|^eat', min=5)
# rstemming.stem('playfullable')
rstemming.stem('eating')
rstemming.stem('eatingness')
```

```python
## Snowball Stemmer
from nltk.stem import SnowballStemmer
snowballsstemmer=SnowballStemmer('english')
```

```python
for word in words:
    print(word+"---->"+snowballsstemmer.stem(word))
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
snowballsstemmer.stem("fairly"),snowballsstemmer.stem("sportingly")
```

```python
snowballsstemmer.stem("goes")
stemming.stem('goes')
# both same 
```

```python
snowballsstemmer.stem("congratulations")
```
