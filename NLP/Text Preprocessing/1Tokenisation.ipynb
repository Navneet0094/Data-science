```python
corpus = """Hello Welcome,to Navneet NLP Tutorials.
Please do watch the entire course! to become expert in NLP.
"""
print(corpus)
```

```python
##  Tokenization
## Sentence-->paragraphs
from nltk.tokenize import sent_tokenize
# sent_tokenize will make sentences when is see a full stop
```

```python
documents =  sent_tokenize(corpus)
print(documents)
type(documents)
```

```python
for sentences in documents:
    print(sentences)
```

```python
# Paragraph -> words
# Sentences -> words  
```

```python
# word tokenize will make it to words whenever it see a comma , fullstop .But not others
# word_puct_tokenize will make to word where it sae a punchation (' , . ) will not cosider as a same word after this ' 
# Treebankwordtokenizer will not consider full stop as word but it will consider last full stop as a word
```

```python
from nltk.tokenize import word_tokenize
# corpus = "Hello Navneet! You're learning NLP. It's fun, isn't it?"
word_tokenize(corpus)
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```
